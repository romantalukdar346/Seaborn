```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
df= sns.load_dataset('flights')
```

```python
df
```

```python
x=df.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True, metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot=True, row_cluster=False,metric='correlation')
```
