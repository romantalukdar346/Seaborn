```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
df= sns.load_dataset('tips')
```

```python
df
```

```python
sns.relplot('total_bill','tip',hue='smoker',style='sex',size='size',sizes=(15,100),data=df)
```

```python
sns.scatterplot('total_bill','tip',hue='smoker',style='sex',size='size',sizes=(15,100),data=df)
plt.legend(loc='upper left')
```

# Categorital Scatter plot
## Strip
## Swarm

```python
sns.catplot(x='day',y='tip',kind='strip',data=df)
```

```python
 sns.catplot(x='day',y='tip',kind='swarm',data=df)
```

```python
 sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=df)
```
