```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np
```

```python
# adding all the dataset

years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
sns.set_style("darkgrid")
```

```python
flower_df =sns.load_dataset("iris")
flower_df
```

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
virginica_df = flower_df[flower_df.species == 'virginica']

from PIL import Image
img=Image.open('chart.jpg')
```

```python
tips_df=sns.load_dataset("tips");
tips_df
```

```python
flights_df=sns.load_dataset("flights");
flights_df
```

```python
df=sns.load_dataset("flights").pivot("month","year","passengers");
df
```

```python
 fig, axes =plt.subplots(2,3)
# for creatimg  sub matrix subplots() used

plt.tight_layout(pad=2)
```

```python
 fig, axes =plt.subplots(2,3, figsize=(12,9))

# for 1st matrix
axes[0,0].plot(years,oranges,'o--r')
axes[0,0].plot(years,apples,'x-g')
axes[0,0].set_title('Yield of Orangens')
axes[0,0].set_xlabel('year')
axes[0,0].set_ylabel('Ton per Hactor')
axes[0,0].legend(["orange","Apple"])

# for 2nd matrix
axes[0,1].set_title('Sepal lenght vs width')
sns.scatterplot(flower_df.sepal_length, 
                flower_df.sepal_width, 
                hue=flower_df.species,
                 s=100, ax= axes[0,1])

# for 3rd matrix
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width,virginica_df.sepal_width],
               bins= np.arange(2,5,0.25),
                stacked=True)
axes[0,2].legend(["setosa","versicolor","virginica"])

# for 4th matrix
axes[1,0].set_title('Restaurant bills')
sns.barplot('day','total_bill',hue='sex',data=tips_df,ax=axes[1,0]);

# for 5th matrix
axes[1,1].set_title('Flight traffic')
sns.heatmap(df,cmap='Blues',ax=axes[1,1])

# for 6th matrix
axes[1,2].set_title('A data science wave')
axes[1,2].grid(False)
axes[1,2].axis('off')
axes[1,2].imshow(img);
# axes[1,2].set_xticks([])
# axes[1,2].set_yticks([])


plt.tight_layout(pad=2);
```
