```python
import matplotlib.pyplot as plt
import seaborn as sns
%malplotlib inline
```

```python
#load data into a pandas dataframe
flower_df =sns.load_dataset("iris")
flower_df
#pre difine dataset in seaborn
```

```python
print(flower_df.species.unique())
```

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width);
```

```python
sns.scatterplot(flower_df.sepal_length,flower_df.sepal_width);
#using scatter plot for better visualization
```

```python
sns.scatterplot(flower_df.sepal_length,
                flower_df.sepal_width,
                hue=flower_df.species,
                s=100);
# hue will give different value different color
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,6))
plt.title("Sepal Dimensions")

sns.scatterplot(flower_df.sepal_length,
                flower_df.sepal_width,
                hue=flower_df.species,
                s=100);
```

```python
plt.figure(figsize=(12,6))
plt.title("Sepal Dimensions")
sns.scatterplot('sepal_length',
                'sepal_width',
                 hue='species',
                 s=100,
                data=flower_df);
# data represent from where data is taken 
```

```python
setosa_df = flower_df[flower_df.species == 'setosa']
versicolor_df = flower_df[flower_df.species == 'versicolor']
virginica_df = flower_df[flower_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha=.4,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=.4,bins=np.arange(2,5,0.25))

plt.legend(['setosa','versicolor']);
#multiple histrograme
```

```python
plt.figure(figsize=(12,8))
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],
         bins=np.arange(2,5,.25),
         stacked=True);
plt.legend(['setosa','versicolor','virginica']);
```

```python
###### Bar chart  ############

years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
plt.plot(years,apples);
```

```python
plt.bar(years,apples);
```

```python
plt.bar(years,apples);
plt.plot(years,apples, 'o--r');
plt.title("Yield of Apples")
```

```python
###strack over bar

plt.bar(years,apples);
plt.bar(years,oranges,bottom=apples)
plt.title("Yield of Apples and Oranges");
plt.legend(['Apples','Oranges'])
```
