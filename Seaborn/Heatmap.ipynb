```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
df= sns.load_dataset('flights')
```

```python
df
```

```python
x=df.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,annot=True,linewidths=0.5,fmt='d')
```

```python
sns.heatmap(x,annot=True,linewidths=0.5,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x,annot=True,linewidths=0.5,fmt='d',cmap='summer')
```
